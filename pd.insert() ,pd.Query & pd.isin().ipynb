```python
import pandas as pd
```

```python
cols= ['GDP',"PCI","Country"]

df =pd.DataFrame([
              [300,20,'Kenya'],
              [150,30,'Uganda'],
              [120,30,'Tanzania'],
              [123,15,'Zambia'],
              [125,32,'Zimbambwe'],
              [130,24,'Somalia'],
              [122,28,'DRC'],
              [131,39,'Rwanda'],
              [115,37,'Burundi']
              ],
              columns = cols ,
             
             )
df
```

```python
df.insert(3,'Rank',[1,2,3,4,5,6,7,8,9])
df
```

```python
Country_Short = pd.Series([])
for i in range(len(df)):
    if df["Country"][i]=="Kenya":
        Country_Short[i] ="KE"
    elif df["Country"][i]=="Uganda":
        Country_Short[i] = "UG"
    elif df["Country"][i]=="Tanzania":
        Country_Short[i] ="TZ"
    elif df["Country"][i]=="Zambia":
        Country_Short[i] ="ZA"
    else:
        Country_Short[i]= df["Country"][i]
df.insert(4,"Country Short",Country_Short)
df
```
