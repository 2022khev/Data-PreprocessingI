```python
import pandas as pd
import numpy as np
```

```python
d= {
    'Name':  ['Alisa','Bobb','Cathrine','Alisa','Bobb',"Cathrine",'Alisa','Bobby','Cathrine'],
    'Subject':['Mathematics','Mathematics','Mathematics','Science','Science','Science','Economics','Economics','Economics'],
    'Score': [62,75,57,78,70,67,89,92,78],
    'Age':[2,23,24,26,26,23,25,27,100]
}
df = pd.DataFrame(d,columns = ['Name','Subject','Score','Age'])
df
```

```python
df['Score_ranked'] = df['Score'].rank(ascending =0)
df
```

```python
df['Score_ranked1'] = df['Score'].rank(ascending =1)
df
```

```python
df['Score_ranked2'] = df['Score'].rank(ascending =0,method ='min')
df
```

```python
df['Score_ranked3'] = df['Score'].rank(ascending =0,method ='max')
df
```

```python
df['Score_ranked4'] = df['Score'].rank(ascending =0,method ='dense')
df
```

```python
df['group_rank'] = df.groupby('Subject')['Score'].rank(ascending=0,method='dense')
df
```

```python
df['group_rank'] = df.groupby('Subject')['Score'].rank(ascending=0,method='dense')
df.set_index(['group_rank'],inplace=True)
df
```

```python
df['dense_rank'] = df['Subject'].rank(ascending=0,method='dense')
df
```
