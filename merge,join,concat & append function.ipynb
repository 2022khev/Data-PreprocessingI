```python
#import libraries
import pandas as pd
import numpy as np
```

```python
#create two dataframes
dict ={'id':[1,2,3,4,5,6,7,8],
    'names' : ["Elv","Jone","Sil","Mary","Agnes","Esther","Philip","Kentucky"],
       'corporate' :[ 'WB','EB','KB','AM','AB',"KEB",'ALB','SB'],
        'amount' :[200,300,354,290,455,267,567,375],
         'type':['proprietor','corporate','partnership','proprietor','proprietor','partnership','corporate','corporate']
            }

dict1 = {'id':[1,2,3,7,8,9,10,8],
         'names' : ["Elv","Jone","Sil","Mary","Agnes","Esther","Philip","Kentucky"],
    'Age' : [26,27,23,34,45,67,44,43],
       'state' :['New York','Mexico','Canada','New York','Mexico','New York','Canada','Mexico'],
        'interest' :[0.5,0.6,0.3,1.55,0.45,0.6,0.45,0.45],
         'type':['proprietor','corporate','partnership','proprietor','proprietor','partnership','corporate','corporate']
            }

df =pd.DataFrame(dict)
df1 = pd.DataFrame(dict1)
print(df)
df1
```

```python
#merge function by default apply inner join
#merge function add columns
pd.merge(df,df1) 
```

```python
#merge function apply inner join based on the id columnns
pd.merge(df,df1,on='id')
```

```python
pd.merge(df,df1, on=['type','names'])
```

```python
#merge function apply inner join based on the id and names columnns
pd.merge(df,df1,on =['id','names'])
```

```python
#suffixes operate on strings columns 
#find matching in both two columns and  return values.
pd.merge(df,df1,suffixes =['_l','_r'],left_on='type',right_on='names')
```

```python
#keep whole wether there is matching or not,otherwise filled with non-values
pd.merge(df,df1,on= 'id',how='outer',suffixes=['_l','_r'])
```

```python
pd.merge(df,df1,on='id',how='outer',suffixes =['_l','_l'])
```

```python
#keep what is shared between both dataframes
pd.merge(df,df1,on= 'id',how='inner',suffixes=['_l','_r'])
```

```python
#keep all the rows from the left dataframe and discard anything that does not match
pd.merge(df,df1,on= 'id',how='left',suffixes=['_l','_r'])
```

```python
#keep all the rows from the right dataframe and discard anything that does not match
pd.merge(df,df1,on= 'id',how='right',suffixes=['_l','_r'])
```
