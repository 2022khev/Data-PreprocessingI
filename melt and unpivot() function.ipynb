```python
df =pd.DataFrame(
{
"Month":['Jan','Feb','Mar','Apri','May','June','Jul','Aug','Sept','Oct','Nov','Dec'],
"2021":[12,-26,-40,10,5,10,2,24,6,4,5,7],
 "2022":[5,4,15,16,14,18,14,8,8,9,9,10],
 "2023":[8,12,18,20,21,8,12,10,11,22,14,12]
 
}
)
df
```

```python
df1 = pd.melt(df,
              id_vars=['Month'],
              value_vars=['2021','2022','2023'],
              var_name = ['Year'],
              value_name='Sales',
              
             )
df1
```

```python
#id_vars , a column that wil not change (s)
#value_vars , are those columns convertted to rows #Can be ignored,=['column1','column2','column3']
#var_name is the  name of column that we store columns that we have convert it to rows,=[name or 'name'
#value_name the values of columns that we have converted them to rows,= "name"

df1 = pd.melt(df,id_vars=['Month'],value_vars=['2021','2022','2023'],var_name =['Year'],value_name='Sales')
df1
```
