```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
import numpy as np
import pandas as pd
%reload_ext autoreload
%autoreload 2
```

```python
x_1 = np.linspace(0,5,10)
y_1 = x_1 **2
plt.plot(x_1,y_1)
plt.title("Days Squared Chart")
plt.ylabel("Days")
plt.ylabel("Days Squared")
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x_1,y_1,'r')
plt.subplot(1,2,2)
plt.plot(x_1,y_1,'b')
```

```python
fig_1 = plt.figure(figsize=(5,4),dpi=100)
axes_1 = fig_1.add_axes([0.1,0.1,0.9,0.9])
axes_1.set_xlabel('Days')
axes_1.set_ylabel('Days Squared')
axes_1.set_title('Days Squared Chart')
axes_1.plot(x_1,y_1,label='x/x^2')
axes_1.plot(y_1,x_1,label='x^2/x')
axes_1.legend(loc=0);
```

```python
fig_1 = plt.figure(figsize=(5,4),dpi=100)
axes_1 = fig_1.add_axes([0.1,0.1,0.9,0.9])
axes_1.set_xlabel('Days')
axes_1.set_ylabel('Days Squared')
axes_1.set_title('Days Squared Chart')
axes_1.plot(x_1,y_1,label='x/x^2')
axes_1.plot(y_1,x_1,label='x^2/x')
axes_1.legend(loc=0);

axes_2 = fig_1.add_axes([0.45,0.45,0.4,0.3])
axes_2.set_xlabel('Days')
axes_2.set_ylabel('Days Squared')
axes_2.set_title('Days Squared Chart')
axes_2.plot(x_1,x_1,'r',label='x^2/x')
```

```python
fig_2,axes_2 = plt.subplots(figsize=(8,4),nrows =1,ncols=3)
plt.tight_layout()
axes_2[0].set_title('Plot 1')
axes_2[1].set_title('Plot 2')
axes_2[2].set_title('Plot 3')
axes_2[1].set_xlabel('x')
axes_2[1].set_ylabel('x squared')
axes_2[0].plot(x_1,y_1,'g')
axes_2[1].plot(x_1,y_1/x_1,'y')
axes_2[2].plot(x_1/y_1,x_1,'r');
```

```python
#unpacking fig,axes from plt.subplots() , figsize =(8,4)
fig_3,axes_3 = plt.subplots(figsize=(8,4),nrows=1,ncols=3)
plt.tight_layout()
axes_3[0].set_title('K')
axes_3[1].set_title('K^2')
axes_3[0].plot(x_1,y_1)
axes_3[1].plot(y_1/x_1,x_1)
axes_3[2].plot(y_1,x_1);
```

```python
df = pd.DataFrame([
              [37,292],
              [40,228],
              [42,320],
              [49,324],
              [53,324],
              [61,376],
              [64,412],
              [72,440],
              [75,496],
              [79,496],
              [81,556],
              [83,536]
    
],columns = ['Tempearature','Sales'])
df
```

```python
np_arr = df.values
x_2 = np_arr[:,0]
y_2 = np_arr[:,1]
fig_5 = plt.figure(figsize=(6,4))
axes_5 = fig_5.add_axes([0,0,1,1])

axes_5.set_xlabel('Temperature')
axes_5.set_ylabel('Ice cream Sales')
axes_5.set_title('Ice cream Sales vs Temperature')
axes_5.plot(x_2,y_2)

axes_5.annotate('Good Month', xy = (83,536),xytext =(60,530),arrowprops = dict(facecolor='red',shrink=0.05));
```
