```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
tips =sns.load_dataset("tips")
```

```python
tips
```

```python
sns.kdeplot(data=tips,
           x="total_bill",
           )
plt.show()
```

```python
iris = sns.load_dataset("iris")
```

```python
sns.kdeplot(data=iris)
plt.show()
```

```python
sns.kdeplot(data=tips,
           x="total_bill",
           bw_adjust=0.2,
           cut=0)
plt.show()
```

```python
sns.kdeplot(data=tips,
           x="total_bill",
           hue="time")
plt.show()
```

```python
sns.kdeplot(data=tips,
           x="total_bill",
           hue="time",
           multiple="stack")
plt.show()
```

```python
diamond = sns.load_dataset("diamonds")
diamond
```

```python
sns.kdeplot(data=diamond,x="price",
           log_scale=True)
plt.show()
```

```python
sns.kdeplot(data=tips,
           x="total_bill",
           hue="size")
plt.show()
```

```python
sns.kdeplot(data=tips,
           x="total_bill",
           hue="size",
            fill=True,
           common_norm=False,
           palette = 'crest',
           alpha=0.5,
           linewidth=0)
plt.show()
```

```python
geyser = sns.load_dataset("geyser")
```

```python
geyser
```

```python
sns.kdeplot(data=geyser,x="waiting",y="duration")
```

```python
sns.kdeplot(data=geyser,
           x="waiting",
           y="duration",
           hue="kind")
plt.show()
```

```python
sns.kdeplot(data=geyser,
           x="waiting",
           y="duration",
           hue="kind",
           fill=True)
plt.show()
```

```python
sns.kdeplot(data=geyser,
           x="waiting",
           y="duration",
           hue="kind",
           levels=5,
           thresh=0.2)
plt.show()
```

```python
sns.kdeplot(data=geyser,
           x="waiting",
           y="duration",
           hue="kind",
           levels=100,
          thresh=0,
          cmap="mako")
plt.show();
```
