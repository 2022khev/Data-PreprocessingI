```python
#The pandas crosstab function is a useful tool for summarizing data. The functionality overlapswith some of the other
# pandas tools but it occuppies a useful place in your data analysis toolbox. 
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame([
     ['Morning','Monday',30],
     ['Morning','Tuesday',36],
     ['Morning','Wednesaday',33],
     ['Morning','Thursday',30],
     ['Morning','Friday',27],
     ['Morning','Saturday',25],
     ['Morning','Sunday',20],
    ['Lunch','Tuesday',35],
    ['Dinner','Wednesday',39],
    ['Lunch','Thursday',40],
    ['Morning','Friday',29],
    ['Lunch','Saturday',34],
    ['Evening','Sunday',35],
    ['Morning','Monday',37],
    ['Lunch','Tuesday',33],
    ['Dinner','Wednesday',38],
    ['Lunch','Thursday',47],
    ['Morning','Friday',45],
    ['Lunch','Saturday',56],
    ['Evening','Sunday',78],
    ['Morning','Monday',39],
    ['Lunch','Tuesday',33],
    ['Dinner','Wednesday',30],
    ['Lunch','Thursday',46],
    ['Morning','Friday',56],
    ['Lunch','Saturday',57],
    ['Evening','Sunday',19]
],
    columns = ['Time','Day','Cost']
)

df.head()
```

```python
pd.crosstab(df['Time'], df['Day'])
```

```python
pd.crosstab(df['Time'], df['Day'], normalize='index', margins=True)
```

```python
pd.crosstab(df['Time'], df['Day'], normalize='columns', margins=True,margins_name='Total')
```

```python
pd.crosstab(df['Time'],df['Day'],values = df.Cost,aggfunc='sum', margins=True,margins_name='Total')
```

```python
pd.crosstab(df['Time'],df['Day'],values = df.Cost,aggfunc='sum',normalize=True, margins=True,margins_name='Total')
```

```python
Type =pd.DataFrame(['Vegatables','Fruits','Proteins','Carbohydrate','Mixed']*4,columns=['Type'])
SupSer = pd.DataFrame(['Sup1','Sup2','Sup3','Sup4','Sup5']*4,columns=['Type'])
df['Type']=Type
df['Sup']=SupSer
df
```

```python
pd.crosstab(df['Time'],[df['Day'],df['Type']])
```

```python
pd.crosstab(df['Time'],[df['Day'],df['Type']])
```

```python
pd.crosstab([df['Time'],df['Sup']],[df['Day'],df['Type']])
```

```python
#virdis #plasma #inferno #magma #cividis #YlGnBu

import seaborn as sns
sns.heatmap(pd.crosstab([df['Time'],df['Sup']],[df['Day'],df['Type']]),
                         cmap = "YlGnBu",annot=True,cbar=False);
```

```python
df1 =pd.DataFrame([
    ['A',3,2,4,8],
    ['B',6,7,3,2],
    ['C',3,3,5,2],
    ['D',1,0,0,0]    
],
    columns= ['Grade','Year1','Year2','Year3','Year4']
 )
df1
```

```python
df_1 = pd.DataFrame([
    
    [3,2,4,8],
    [6,7,3,2],
    [3,3,5,2],
    [1,0,0,0] 
],
columns= ['Year1','Year2','Year3','Year4'],
index = ['A','B','C','D']  
 )
df_1.columns.name='Year'
df_1.index.name='Grade'
df_1
```

```python
df_1['Total_Grade'] =pd.DataFrame(df_1.sum(axis=1)).astype(int)
df_1
```

```python
df2 = pd.melt(df1,id_vars='Grade',var_name='Year',value_name='Count')
df2
```

```python
pd.crosstab(df2.Grade,df2.Year,values= df2.Count ,aggfunc='sum',margins=True,margins_name='Total')
```

```python
sns.heatmap(pd.crosstab(df2['Grade'],df2['Year']),
                         cmap = "YlGnBu",annot=True,cbar=False);
```
