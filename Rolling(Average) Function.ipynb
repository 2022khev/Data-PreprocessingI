```python
ind = pd.date_range(start = '2021-01-01',end='2022-08-23', freq='d').to_list()
cols = ['Open','High','Low','Close','Volume']
df =pd.DataFrame(np.random.rand(600,5),index=ind,columns=cols)
df.index.names=['Date']
df.head()
```

```python
df.tail()
```

```python
df['MV1'] = df.Close.rolling(9).mean()
df.head(10)
```

```python
df['MV2'] = df.Close.rolling(21).mean()
df.head(10)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15,10))
plt.grid(True)
plt.plot(df['Close'],label="GLD")
plt.plot(df['MV1'],label="Moving Average 9")
plt.plot(df['MV2'],label="Moving Average 21")
plt.legend(loc=2);
```

```python
df['Change'] = np.log(df['Close']/df['Close'].shift())
```

```python
plt.figure(figsize=(20,7))
plt.plot(df['Change']);
```

```python
df['Volatility'] = df.Change.rolling(21).std().shift()
df["Volatility"].plot();
```

```python
ma =21
df['returns'] = np.exp(df['Close'].diff())
df['ma'] = df['Close'].rolling(ma).mean()
df['ratio'] =df['Close']/df['ma']
```

```python
df['Close'].describe()
```

```python
percentiles = [5,10,50,90,90]
p=np.percentile (df['ratio'].dropna(),percentiles)
```

```python
plt.figure(figsize=(15,7))

df['ratio'].dropna().plot(legend=True)
plt.axhline(p[0],c=(.5,.5,.5),ls='--')
plt.axhline(p[2],c=(.5,.5,.5),ls='--')
plt.axhline(p[-1],c=(.5,.5,.5),ls='--');
```

```python
short = p[-1]
long = p[0]
df['position'] = np.where(df.ratio > short, -1,np.nan)
df['position'] = np.where(df.ratio  < long, 1,df['position'])
df['position']= df['position'].ffill()
```

```python
plt.figure(figsize=(15,7))
df['position'].dropna().plot();
```

```python
df['strat_return'] =df['returns'] * df['position'].shift()
```

```python
plt.plot(np.exp(df['returns']).dropna().cumprod(),label='Buy/Hold')
plt.plot(np.exp(df['strat_return']).dropna().cumprod(),label='Strategy')
plt.legend();
```
